```python
import requests

api_key = "4c54626d6c6a776b31334f6e686265"
base_url = "http://openapi.seoul.go.kr:8088"
# /{api_key}/json/energyUseDataSummaryInfo/{pageNo}/{pageNo + numOfRows - 1}/{year}/{month}

filtered_data = []
beg_year = 2015
end_year = 2024
beg_month = 1
end_month = 12
numOfRows = 1000

for year in range(beg_year, end_year + 1):
    for month in range(beg_month, end_month + 1):
        pageNo = 1
        while True:
            url = f"{base_url}/{api_key}/json/energyUseDataSummaryInfo/{pageNo}/{pageNo + numOfRows - 1}/{year}/{month:02d}"

            response = requests.get(url)

            if response.status_code == 200:
                print(f"api 호출 성공 (년도: {year}, 월: {month}, 페이지: {pageNo})")
                try:
                    data = response.json()
                    energy_data = data.get('energyUseDataSummaryInfo')

                    if energy_data and 'row' in energy_data:
                        rows = energy_data['row']
                        # MM_TYPE == "개인"
                        for row in rows:
                            if row.get('MM_TYPE') == '개인':
                                filtered_row = {
                                    '년도': row.get('YEAR'),
                                    '월': row.get('MON'),
                                    '현년 전기 사용량': row.get('EUS'),
                                    '현년 가스 사용량': row.get('GUS'),
                                    '현년 수도 사용량': row.get('WUS'),
                                    '현년 지역난방 사용량': row.get('HUS')
                                }
                                filtered_data.append(filtered_row)

                        if len(rows) < numOfRows:
                            break
                        else:
                            pageNo += numOfRows

                    elif energy_data and energy_data.get('RESULT') and energy_data['RESULT'].get('CODE') == 'INFO-200':
                         print(f"{year}년 {month}의 데이터가 없습니다.")
                         break
                    else:
                        print(f"'row' 데이터가 없거나 예상치 못한 오류: {data}")
                        break

                except requests.exceptions.JSONDecodeError as e:
                    print(f"JSON 디코딩 오류 (년도: {year}, 월: {month}, 페이지: {pageNo}): {e}")
                    break
            else:
                print(f"API 호출 실패 (년도: {year}, 월: {month}, 페이지: {pageNo}): {response.status_code}")
                break


print("\n2015년 1월부터 2024년 12월까지의 개인 유형의 현년 전기, 가스, 수도, 지역난방 에너지 사용랑 데이터:")
print(filtered_data)
print(f"데이터 개수: {len(filtered_data)}")

# To convert to pandas DataFrame later:
# import pandas as pd
# df_personal_data = pd.DataFrame(filtered_data)
# display(df_personal_data)
```
